```python
city_day = city_day_df.dropna(subset=["AQI_Bucket"])
city_day
```

```python
city_day["Date"] = pd.to_datetime(city_day["Date"])
city_day["Year"] = city_day["Date"].dt.year
```

```python
bad = ["Poor", "Very Poor", "Severe"]
bad_air = city_day[city_day["AQI_Bucket"].isin(bad)]
bad_air_yearly = (
    bad_air.groupby(["City", "Year"])["AQI_Bucket"]
    .count()
    .reset_index(name="Bad_Air_Days")
)
```

```python
bad_air_yearly
```

```python
total_days_yearly = (
    city_day.groupby(["City", "Year"])["AQI_Bucket"]
    .count()
    .reset_index(name="Total_Valid_Days")
)
merged = bad_air_yearly.merge(total_days_yearly, on=["City", "Year"])
merged["Bad_Air_Percentage"] = (
    merged["Bad_Air_Days"] / merged["Total_Valid_Days"]
)
merged
```

```python
city_rank = merged.groupby("City")["Bad_Air_Percentage"].mean().sort_values(ascending=False)
top_cities = city_rank.head(5).index
```

```python
top_cities
```

```python
top_cities = ['Ahmedabad', 'Delhi', 'Patna', 'Gurugram', 'Lucknow']
df_top = merged[merged['City'].isin(top_cities)].copy()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
for city in top_cities:
    df_city = df_top[df_top['City'] == city]
    fig.add_trace(go.Scatter(
        x=df_city["Year"],
        y=df_city["Bad_Air_Percentage"],
        mode="lines+markers",
        name=city,
        visible=True 
    ))
buttons = []
for i, city in enumerate(top_cities):
    visibility = [False] * len(top_cities)
    visibility[i] = True  
    
    buttons.append(dict(
        label=city,
        method="update",
        args=[{"visible": visibility},
              {"title": f"Bad-Air Percentage Trend: {city}"}]
    ))
buttons.insert(0, dict(
    label="Show All",
    method="update",
    args=[{"visible": [True] * len(top_cities)},
          {"title": "Yearly Bad-Air Percentage for All Top Cities"}]
))
fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=1.15
    )],
    title="Yearly Bad-Air Percentage for Top Polluted Cities",
    xaxis_title="Year",
    yaxis_title="Bad-Air Percentage",
    legend=dict(
    orientation="h",
    y=-0.2
)
)
fig.show()
```

### Inference:-

To compare cities fairly, we first removed days with missing AQI, then classified each day into AQI buckets. “Bad-air days” were defined as days with AQI in Poor, Very Poor or Severe. For each City–Year pair, we calculated:

Bad-Air Percentage = Bad-Air Days / Total Valid AQI Days

Next, we ranked cities by their average Bad-Air Percentage across 2015–2020 and selected the top 5 (Ahmedabad, Delhi, Patna, Gurugram, Lucknow). The line chart shows how the fraction of heavily polluted days has changed over time for these worst-affected cities.

Drop in Bad Air Percentage Days:-

In simple words — the number of very polluted days reduced compared to before.
It does not mean the air became clean; it only means the worst days happened less often.

The plot suggests that while all five cities remain highly polluted, most of them show a gradual decline in the share of bad-air days, with a sharp drop in 2020 (partly due to COVID-19 lockdowns and reduced activity). This indicates some reduction in extreme pollution episodes, not that the cities have become clean overall.

# Seasonal Effects:-

```python
city_day_df['Date'] = pd.to_datetime(city_day_df['Date'])
city_day_df['Year'] = city_day_df['Date'].dt.year
city_day_df['Month'] = city_day_df['Date'].dt.month
city_day_df['MonthName'] = city_day_df['Date'].dt.month_name()
city_day = city_day_df.dropna(subset=['AQI_Bucket'])
city_day
```

```python
monthly_aqi = (
    city_day.groupby(['Year','Month'])['AQI']
    .mean()
    .reset_index()
)
```

```python
monthly_aqi
```

```python
heatmap_df = monthly_aqi.pivot(index='Year', columns='Month', values='AQI')
```

```python
import plotly.express as px

fig = px.imshow(
    heatmap_df,
    labels=dict(x="Month", y="Year", color="Avg AQI"),
    x=[ 'Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec' ],
    title="Monthly Average AQI Heatmap (All Cities Combined)",
    color_continuous_scale='RdYlGn_r'
)

fig.update_layout(height=600)
fig.show()
```

### Inference:-
The heatmap highlights seasonal spikes very clearly. Darker colors in October to January show heavy pollution in winter, while lighter colors in June–August show cleaner air during the monsoon. This pattern repeats every year, proving that India’s air quality is strongly seasonal. It also becomes easy to compare how severe winter pollution was across different years.

```python
fig = go.Figure()
years = sorted(monthly_aqi['Year'].unique())
for yr in years:
    data = monthly_aqi[monthly_aqi['Year'] == yr]
    fig.add_trace(go.Scatter(
        x=data['Month'],
        y=data['AQI'],
        fill='tozeroy',
        mode='lines+markers',
        name=str(yr),
        visible=True if yr == years[0] else False   # show first year initially
    ))
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(
                    label=str(yr),
                    method='update',
                    args=[
                        {"visible": [yr == y for y in years]},
                        {"title": f"Monthly AQI Area Chart for {yr}"}
                    ]
                ) for yr in years
            ],
            direction="down",
            showactive=True,
            x=1.15,
            y=1
        )
    ],
    title=f"Monthly AQI Area Chart for {years[0]}",
    xaxis=dict(
        tickmode='array',
        tickvals=list(range(1,13)),
        ticktext=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    ),
    yaxis_title="Average AQI",
    xaxis_title="Month",
    height=500
)
fig.show()
```

### Inference:-
This graph shows how the average AQI changes month-wise across India for each year. A clear seasonal pattern appears in every year. AQI is lowest during the monsoon months (June–August), when rainfall helps wash out pollutants from the air. After the monsoon ends, AQI starts rising again and reaches its peak between October and January. This spike happens because of a mix of winter weather conditions that trap pollutants near the ground, along with events like crop-burning in northern states and Diwali fireworks.

Most years follow the same curve:-
- Monsoon → Cleanest air
- Post-monsoon + Winter → Worst air

This confirms that India’s air pollution follows a strong seasonal cycle driven by both climate and human activities.

- Winter Effect (Weather Impact):-

In winter (Nov–Jan), temperatures drop and the wind becomes very weak. Because of this, polluted air gets trapped close to the ground instead of spreading out. So even normal levels of pollution look much worse. That’s why AQI is always highest during these months.

- Monsoon Effect (Natural Cleaning):-

During June–August, rainfall washes away dust and fine particles like PM2.5 and PM10. This naturally brings down pollution levels across the country. This is why monsoon months always show the cleanest air.

- Crop-Burning Effect (Human Activity):-

The spike in October–November matches the crop-burning season in Punjab, Haryana, and Uttar Pradesh. Burning crop residue releases a lot of PM2.5, PM10, CO, and NOx. Cities in North India (like Delhi, Gurugram, Lucknow, Patna) show a sharp jump in AQI during this time, while South Indian cities stay more stable. So the early-winter pollution spike happens because of extra emissions from crop burning and winter weather that traps these pollutants.

- Summary:-

India’s air quality changes mainly due to weather patterns and human activities.
The repeating yearly cycle — winter spike, monsoon drop — shows the impact of weather.
The big jump in October–November highlights the contribution of crop burning.

# Festival Effects:-

```python
festival_dates = {
    "Diwali": {
        2015: "2015-11-11",
        2016: "2016-10-30",
        2017: "2017-10-19",
        2018: "2018-11-07",
        2019: "2019-10-27",
        2020: "2020-11-14"
    },
    "Dussehra": {
        2015: "2015-10-22",
        2016: "2016-10-11",
        2017: "2017-09-30",
        2018: "2018-10-19",
        2019: "2019-10-08",
        2020: "2020-10-25"
    },
    "Holi": {
        2015: "2015-03-06",
        2016: "2016-03-24",
        2017: "2017-03-13",
        2018: "2018-03-02",
        2019: "2019-03-21",
        2020: "2020-03-10"
    }
}
```

```python
def assign_festival_label(row):
    year = row['Year']
    date = row['Date']
    for fest, fest_years in festival_dates.items():
        fest_date = pd.to_datetime(fest_years.get(year, None))
        if pd.isna(fest_date):
            continue
        before_window = (fest_date - pd.Timedelta(days=7))
        after_window  = (fest_date + pd.Timedelta(days=7))
        if date == fest_date:
            return fest + "_Day"
        elif before_window <= date < fest_date:
            return fest + "_Before"
        elif fest_date < date <= after_window:
            return fest + "_After" 
    return "None"
```

```python
city_day['Festival'] = city_day.apply(assign_festival_label, axis=1)
diwali_df = city_day[city_day['Festival'].str.contains("Diwali")]
holi_df = city_day[city_day['Festival'].str.contains("Holi")]
dussehra_df = city_day[city_day['Festival'].str.contains("Dussehra")]
```

```python
diwali_summary = (
    diwali_df.groupby('Festival')['AQI']
    .mean()
    .reset_index()
)

holi_summary = (
    holi_df.groupby('Festival')['AQI']
    .mean()
    .reset_index()
)

dussehra_summary = (
    dussehra_df.groupby('Festival')['AQI']
    .mean()
    .reset_index()
)
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=diwali_summary['Festival'],
    y=diwali_summary['AQI'],
    name='Diwali',
    marker_color='blue',
    visible=True 
))
fig.add_trace(go.Bar(
    x=holi_summary['Festival'],
    y=holi_summary['AQI'],
    name='Holi',
    marker_color='green',
    visible=False
))

fig.add_trace(go.Bar(
    x=dussehra_summary['Festival'],
    y=dussehra_summary['AQI'],
    name='Dussehra',
    marker_color='red',
    visible=False
))
fig.update_layout(
    updatemenus=[
        dict(
            type="dropdown",
            x=1.15, y=1,
            showactive=True,
            buttons=[
                dict(
                    label="Diwali",
                    method="update",
                    args=[{"visible": [True, False, False]},
                          {"title": "AQI Impact Around Diwali (±7 Days)"}]
                ),
                dict(
                    label="Holi",
                    method="update",
                    args=[{"visible": [False, True, False]},
                          {"title": "AQI Impact Around Holi (±7 Days)"}]
                ),
                dict(
                    label="Dussehra",
                    method="update",
                    args=[{"visible": [False, False, True]},
                          {"title": "AQI Impact Around Dussehra (±7 Days)"}]
                )
            ]
        )
    ],
    title="Festival-wise AQI Impact (Select a Festival)",
    xaxis_title="Festival Window",
    yaxis_title="Average AQI",
    height=500
)
fig.show()
```

### Inference:-
- This analysis shows how AQI behaves around 3 festivals each year. By tagging each date into before, during, and after windows (±7 days), we can separate normal pollution levels from festival-related changes. We observe a clear jump in AQI on Diwali day and in the following days, indicating the impact of fireworks combined with the winter inversion period. This confirms that Diwali consistently contributes to short-term spikes in PM2.5 and overall air quality degradation.

- AQI is not drastically different across “Before”, “Festival Day”, and “After” windows for Holi and Dussehra. The values stay within a similar range, which suggests that these festivals do not cause a strong or sudden nationwide pollution spike.

- Dussehra shows slightly higher AQI after the festival, but the difference is small. This pattern could come from regular seasonal changes (late-October winter buildup) rather than the festival itself.

- Holi shows almost no change across the three windows. This is expected because Holi celebrations don’t involve activities that add large amounts of particulate matter.

- These results do NOT mean the festivals have zero pollution impact. They only mean that when we take a country-wide average, the effect is too diluted to be visible clearly. City-level spikes might still exist, but they get averaged out in national data.

# Hourly Analysis:-

```python
city_hour = city_hour_df.dropna(subset=['AQI_Bucket'])
city_hour.info()
```

```python
df = city_hour[city_hour['City'].isin(['Delhi', 'Mumbai'])].copy()
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Date'] = df['Datetime'].dt.date
df['Hour'] = df['Datetime'].dt.hour
df['Year'] = df['Datetime'].dt.year
```

```python
diwali_dates = {
    2015: '2015-11-11',
    2016: '2016-10-30',
    2017: '2017-10-19',
    2018: '2018-11-07',
    2019: '2019-10-27',
    2020: '2020-11-14'
}
```

```python
fig = go.Figure()

cities = ['Delhi', 'Mumbai']
years = [2018, 2019]

initial_city = 'Delhi'
initial_year = 2019
diwali_day = pd.to_datetime(diwali_dates[initial_year]).date()
normal_day = (pd.to_datetime(diwali_dates[initial_year]) - pd.Timedelta(days=7)).date()

df_city_year = df[(df['City'] == initial_city) & (df['Year'] == initial_year)]

dd = df_city_year[df_city_year['Date'] == diwali_day]
nd = df_city_year[df_city_year['Date'] == normal_day]

fig.add_trace(go.Scatter(
    x=dd['Hour'], y=dd['AQI'], mode='lines+markers',
    name='Diwali Day', line=dict(color='red')
))
fig.add_trace(go.Scatter(
    x=nd['Hour'], y=nd['AQI'], mode='lines+markers',
    name='Normal Day', line=dict(color='blue')
))
buttons = []
for city in cities:
    for year in years:

        diwali = pd.to_datetime(diwali_dates[year]).date()
        normal = (pd.to_datetime(diwali_dates[year]) - pd.Timedelta(days=7)).date()

        temp = df[(df['City'] == city) & (df['Year'] == year)]

        dd2 = temp[temp['Date'] == diwali]
        nd2 = temp[temp['Date'] == normal]

        buttons.append(dict(
            label=f"{city} - {year}",
            method="update",
            args=[
                {
                    "x": [dd2['Hour'], nd2['Hour']],
                    "y": [dd2['AQI'], nd2['AQI']]
                },
                {
                    "title": f"Hourly AQI on Diwali vs Normal Day — {city} ({year})"
                }
            ]
        ))

fig.update_layout(
    updatemenus=[{
        "buttons": buttons,
        "direction": "down",
        "showactive": True,
        "x": 1.15,
        "y": 1
    }],
    title=f"Hourly AQI on Diwali vs Normal Day — {initial_city} ({initial_year})",
    xaxis_title="Hour of Day",
    yaxis_title="AQI",
    height=550,
    legend=dict(
    orientation="h",
    y=-0.2
))
fig.show()
```

### Inference:-
This graph compares hourly AQI on Diwali with a normal winter day for different years.
The trend is very clear:

- Delhi shows a heavy spike between 7 PM and 2 AM on Diwali night.
This comes from fireworks + winter inversion, which traps pollutants near the ground.
Even if the day remains normal, the AQI shoots up at night, and this pattern repeats almost every year.

- Mumbai, on the other hand, stays almost flat on both days.
There is little or no Diwali spike.
The sea breeze and coastal humidity help disperse pollution quickly, so the effect of fireworks doesn’t accumulate in the air.

Overall, Delhi shows a strong Diwali impact while Mumbai barely reacts.
The difference highlights how geography + weather control how festivals affect pollution.

- The coastal wind keeps air moving, so pollution doesn’t build up at specific times.
- The morning and evening peaks are present but very small.

This comparison shows that Delhi’s pollution is heavily influenced by local emissions + atmospheric trapping, while Mumbai benefits from continuous natural ventilation.

# Pollutant Analysis:-

```python
df = city_day_df.copy()

df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

pollutants = ['PM2.5', 'PM10', 'NO2', 'NOx', 'SO2', 'CO', 'O3']

monthly_pollutants = df.groupby('Month')[pollutants].mean().reset_index()


fig = go.Figure()
for pollutant in pollutants:
    fig.add_trace(go.Scatter(
        x=monthly_pollutants['Month'],
        y=monthly_pollutants[pollutant],
        mode='lines+markers',
        name=pollutant
    ))
fig.update_layout(
    title="Monthly Average Concentration of Pollutants (All India)",
    xaxis_title="Month",
    yaxis_title="Average Pollutant Concentration",
    xaxis=dict(tickmode='array', tickvals=list(range(1,13))),
    height=600
)
fig.show()
```

### Inference:-
When we look at the monthly pollutant trends across all cities, a clear seasonal pattern appears:

- PM2.5 and PM10 shoot up in October–November, which matches the crop-burning season in North India.
- They remain high in December–January because winter traps pollution near the ground (temperature inversion).
- Ozone (O₃) is highest in summer, since strong sunlight increases ozone formation.
- Monsoon months (June–August) show the lowest values for almost every pollutant because rain cleans the air.

Overall, this graph shows that both weather and specific human activities (crop burning, winter stagnation) strongly shape pollution levels across India.

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'
```

```python
df['Season'] = df['Month'].apply(get_season)
pollutants = ['PM2.5', 'PM10', 'NO2', 'NOx', 'SO2', 'CO', 'O3']
seasonal = df.groupby('Season')[pollutants].mean().reset_index()
season_order = ['Winter', 'Summer', 'Monsoon', 'Post-Monsoon']
seasonal = seasonal.set_index('Season').loc[season_order].reset_index()
```

```python
fig = go.Figure()

for pollutant in pollutants:
    fig.add_trace(go.Bar(
        x=seasonal['Season'],
        y=seasonal[pollutant],
        name=pollutant
    ))

fig.update_layout(
    title="Seasonal Comparison of Pollutants (All India)",
    xaxis_title="Season",
    yaxis_title="Average Concentration",
    barmode='group',
    height=600
)

fig.show()
```

### Inference:-
This seasonal comparison makes the pollution patterns very clear:
- Winter has the highest PM2.5 and PM10, which matches the winter inversion effect — cold, still air traps pollutants close to the ground.
- Post-monsoon (Sep–Nov) also shows a sharp rise in PM levels, mainly because of crop burning in Punjab, Haryana, and UP.
- During Diwali (also in Oct–Nov), fireworks release a lot of PM2.5, PM10, SO₂, and NOx, which adds to the pollution spike already caused by crop burning.
- Summer shows higher O₃, because strong sunlight converts NOx into ozone.
- Monsoon has the lowest pollution for almost every pollutant, since rain naturally washes the air clean.

```python
df = city_day.copy()
df["Date"] = pd.to_datetime(df["Date"])
pollutants = ["PM2.5", "PM10", "NO2", "NOx", "SO2", "CO"]
city_pollutants = df.groupby("City")[pollutants].mean().reset_index()
city_pollutants
```

```python
#to fill lucknow and jorakhpur nan values of pm10 and nox respectively
city_pollutants = city_pollutants.fillna(city_pollutants.median(numeric_only=True))
```

```python
city_pollutants
```

```python
exposure = (
    df.assign(unhealthy = df["AQI"] > 100)
      .groupby("City")["unhealthy"]
      .mean()
      .reset_index()
      .rename(columns={"unhealthy": "ExposureFactor"})
)
exposure
```

```python
norm_df = city_pollutants.copy()
for col in pollutants:
    min_val = norm_df[col].min()
    max_val = norm_df[col].max()
    norm_df[col] = (norm_df[col] - min_val) / (max_val - min_val + 1e-8)
```

```python
weights = {
    "PM2.5": 0.40,
    "PM10": 0.20,
    "NO2": 0.15,
    "NOx": 0.10,
    "SO2": 0.10,
    "CO": 0.05
}
norm_df["WeightedRisk"] = sum(norm_df[col] * w for col, w in weights.items())
risk_df = norm_df.merge(exposure, on="City", how="left")
risk_df["FinalRiskScore"] = risk_df["WeightedRisk"] * (1 + risk_df["ExposureFactor"])
risk_df = risk_df.sort_values("FinalRiskScore", ascending=False)
risk_df
```

```python
def categorize(score):
    if score > 1.4: return "Severe"
    if score > 1.0: return "High"
    if score > 0.6: return "Moderate"
    return "Low"

risk_df["RiskCategory"] = risk_df["FinalRiskScore"].apply(categorize)
```

# Station Performance and Data Quality Audit:-

```python
df = station_hour_df.copy()
df['Datetime'] = pd.to_datetime(df['Datetime'])
pollutants = [
    'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
    'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI'
]
```

```python
base = (
    df.groupby('StationId')['Datetime']
      .agg(start_dt='min', end_dt='max', actual_hours='count')
)
```

```python
time_diff_hours = (base['end_dt'] - base['start_dt']).dt.total_seconds() / 3600
base['expected_hours'] = time_diff_hours.astype(int) + 1 
base['coverage_pct'] = base['actual_hours'] / base['expected_hours']
base['missing_pct'] = 1 - base['coverage_pct']
base = base[base['expected_hours'] > 0]
base
```

```python
pollutant_counts = (
    df.groupby('StationId')[pollutants]
      .count()   
)
pollutant_cov = pollutant_counts.div(base['expected_hours'], axis=0)
pollutant_cov = pollutant_cov.clip(upper=1.0) 
pollutant_cov = pollutant_cov.add_suffix('_cov')
station_quality = base.join(pollutant_cov)
station_quality = station_quality.sort_values('coverage_pct', ascending=False)
```

```python
subset = station_quality.sort_values('coverage_pct', ascending=True)
poll_cov_cols = [c for c in station_quality.columns if c.endswith('_cov')]
heat_df = subset[poll_cov_cols].copy()
heat_df.index.name = 'StationId'
heat_df = heat_df * 100
fig = px.imshow(
    heat_df,
    labels=dict(x="Pollutant", y="Station", color="Coverage (%)"),
    aspect='auto',
    text_auto='.1f',
    title="Pollutant-wise Coverage per Station"
)
fig.update_layout(height=)
fig.show()
```

```python
poll_cov_cols = [c for c in station_quality.columns if c.endswith('_cov')]
station_quality['pollutant_coverage_mean'] = station_quality[poll_cov_cols].mean(axis=1)
station_quality['reliability_score'] = (
    (1 - station_quality['missing_pct']) * 0.4 +
    station_quality['pollutant_coverage_mean'] * 0.6
)
```

```python
def reliability_category(score):
    if score >= 0.85:
        return "High"
    elif score >= 0.70:
        return "Moderate"
    else:
        return "Low"
```

```python
station_quality['reliability_category'] = station_quality['reliability_score'].apply(reliability_category)
top_reliable = station_quality.sort_values('reliability_score', ascending=False).head(10)
bottom_reliable = station_quality.sort_values('reliability_score', ascending=True).head(10)
top_reliable = top_reliable[['reliability_score', 'reliability_category', 'coverage_pct', 'pollutant_coverage_mean']]
bottom_reliable = bottom_reliable[['reliability_score', 'reliability_category', 'coverage_pct', 'pollutant_coverage_mean']]
```

```python
fig_top = px.bar(
    top_reliable.reset_index(),
    x='StationId',
    y='reliability_score',
    title="Top 10 Most Reliable Stations",
    labels={'reliability_score': 'Reliability Score'}
)
fig_top.update_layout(xaxis_tickangle=-45)
fig_top.show()
```

```python
fig_bottom = px.bar(
    bottom_reliable.reset_index(),
    x='StationId',
    y='reliability_score',
    title="Bottom 10 Least Reliable Stations",
    labels={'reliability_score': 'Reliability Score'}
)
fig_bottom.update_layout(xaxis_tickangle=-45)
fig_bottom.show()
```
